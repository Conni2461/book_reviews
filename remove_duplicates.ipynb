```python
from sqlalchemy import create_engine, Column, Date, String, Integer, Float, exc, and_, or_
from sqlalchemy.orm import Session, declarative_base

import requests

from selectorlib import Extractor
import requests
import json
from time import sleep
import random
import re
```

```python
engine = create_engine("sqlite+pysqlite:///books.sqlite")
Base = declarative_base()
Base.metadata.create_all(engine)
session = Session(engine)
```

```python
class merged_books(Base):
    __tablename__ = "merged_books"

    isbn = Column(String, primary_key=True, nullable=False)
    title = Column(String, nullable=True)
    author = Column(String, nullable=True)
    publisher = Column(String, nullable=True)
    published_date = Column(String, nullable=True)
    description = Column(String, nullable=True)
    maturity_rating = Column(Integer, nullable=True)
    
    page_count = Column(Integer, nullable=True)
    score_nyt = Column(Integer, nullable=True)
    goodreads_rating = Column(Float, nullable=True)
    goodreads_count = Column(Integer, nullable=True)
    amazon_rating = Column(Float, nullable=True)
    amazon_count = Column(Integer, nullable=True)
    google_rating = Column(Float, nullable=True)
    google_count = Column(Integer, nullable=True)

    book_image = Column(String, nullable=True)
    amazon_link = Column(String, nullable=True)
```

```python
books = session.query(merged_books).all()
```

```python
finished_books = set()
N = 3


def similarity(set1:set, set2:set, thresh = 0.8):
    return (len(set1.intersection(set2)) / len(set1.union(set2))) > thresh

for book in books:
    if book in finished_books:
        continue
    #print(f"{book.title}, {book.author}")
    grams_title = set([book.title[i:i+N] for i in range(len(book.title)-N+1)])
    grams_author = set([book.author[i:i+N] for i in range(len(book.author)-N+1)])
    
    for book2 in books:
        grams_title2 = set([book2.title[i:i+N] for i in range(len(book2.title)-N+1)])
        grams_author2 = set([book2.author[i:i+N] for i in range(len(book2.author)-N+1)])

        if similarity(grams_title, grams_title2) & similarity (grams_author, grams_author2, 0.5):
            print(f"{book.title}, {book.author}")
            print(f"{book2.title}, {book2.author}")

            finished_books.add(book2)
        



```
